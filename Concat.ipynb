```python
import pandas as pd

def standardize_column_names(df):
    """
    Standardize column names using an expanded mapping dictionary.
    """
    # Create a mapping for different variations of column names
    column_mapping = {
        # Sheet title variations
        'coverage sheeet': 'coverage sheet',
        'coverage sheet title': 'coverage sheet',
        'spread point coverage sheet  -': 'coverage sheet',
        'metal master coverage sheet sets - 24012020': 'coverage sheet',
        'machine guard coverage sheet sets - 14052020': 'coverage sheet',
        'metso coverage sheet - 23102023': 'coverage sheet',
        
        # Publication number variations
        'publication number / i.d number': 'publication/id number',
        'publication number': 'publication/id number',
        
        # Program variations
        'relavant program': 'relevant programme',
        'program covered': 'relevant programme',
        'relevant program': 'relevant programme',
        
        # Coverage level variations
        'levels': 'coverage levels available',
        
        # Date/Check variations
        'date last checked / amended': 'checked',
        
        # Unnamed columns mapping
        'unnamed: 1': 'checked',
        'unnamed: 2': 'amended',
        'unnamed: 3': 'coverage levels available',
        'unnamed: 4': 'publication/id number',
        'unnamed: 5': 'relevant programme'
    }
    
    # Convert all column names to lowercase and strip whitespace
    df.columns = df.columns.str.lower().str.strip()
    
    # Rename columns based on mapping
    renamed_columns = {}
    for col in df.columns:
        if col in column_mapping:
            renamed_columns[col] = column_mapping[col]
    
    if renamed_columns:
        df = df.rename(columns=renamed_columns)
    
    # Special handling for sheets with combined date checked/amended column
    if 'date last checked / amended' in df.columns:
        df['checked'] = df['date last checked / amended']
        df = df.drop('date last checked / amended', axis=1)
    
    # If 'checked' exists but 'amended' doesn't, copy 'checked' to 'amended'
    if 'checked' in df.columns and 'amended' not in df.columns:
        df['amended'] = df['checked']
    
    return df

def combine_excel_sheets_vertically(excel_file, sheet_numbers=None):
    """
    Combines selected sheets from an Excel file vertically (row-wise),
    standardizing column names before combining.
    """
    # Read all sheet names
    excel = pd.ExcelFile(excel_file)
    all_sheets = excel.sheet_names
    
    # If sheet_numbers not specified, use all sheets except the first one
    if sheet_numbers is None:
        sheet_numbers = list(range(1, len(all_sheets)))
    
    # Initialize an empty list to store DataFrames
    dfs = []
    
    # Expected columns after standardization
    expected_columns = [
        'amended', 'checked', 'coverage levels available',
        'coverage sheet', 'publication/id number', 'relevant programme'
    ]
    
    # Process each sheet
    successful_sheets = 0
    for sheet_num in sheet_numbers:
        if sheet_num < len(all_sheets):
            try:
                # Read the sheet into a DataFrame
                df = pd.read_excel(excel_file, sheet_name=sheet_num)
                
                # Skip empty sheets
                if df.empty:
                    print(f"Skipping empty sheet: {all_sheets[sheet_num]}")
                    continue
                
                # Standardize column names
                df = standardize_column_names(df)
                
                # Remove any comment columns
                if 'comment' in df.columns:
                    df = df.drop('comment', axis=1)
                
                # Add any missing columns with NULL values
                for col in expected_columns:
                    if col not in df.columns:
                        df[col] = None
                
                # Keep only expected columns
                df = df[expected_columns]
                
                # Add source sheet information
                df['source_sheet'] = all_sheets[sheet_num]
                
                # Append to the list
                dfs.append(df)
                successful_sheets += 1
                print(f"Successfully processed sheet: {all_sheets[sheet_num]}")
                
            except Exception as e:
                print(f"Error reading sheet '{all_sheets[sheet_num]}': {str(e)}")
    
    # Combine all DataFrames vertically
    if dfs:
        combined_df = pd.concat(dfs, ignore_index=True)
        # Ensure consistent column order
        final_columns = expected_columns + ['source_sheet']
        combined_df = combined_df[final_columns]
        print(f"\nSuccessfully combined {successful_sheets} sheets.")
        return combined_df
    else:
        print("No sheets were successfully processed.")
        return pd.DataFrame()  # Return empty DataFrame if no sheets were combined

df = combine_excel_sheets_vertically(file_path)
print("\nCombined DataFrame Info:")
print(df.info())
```
